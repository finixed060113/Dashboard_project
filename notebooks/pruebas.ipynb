```python
import requests

planet_names = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune"]

planet_data = {}  # Dictionary to store planet data

for planet_name in planet_names:
    
    api_url = f"https://api.le-systeme-solaire.net/rest/bodies/{planet_name}"
    
    response = requests.get(api_url)
    if response.status_code == 200:
        planet_info = response.json()
        planet_data[planet_name] = planet_info
    else:
        print(f"Error fetching data for {planet_name}")

print(planet_data)  # Display the collected planet data
```

```python
planet_data
```

```python
planets = {
    'name': [],
    'semimajorAxis': [],
    'perihelion': [],
    'aphelion': [],
    'eccentricity': [],
    'inclination': [],
    'mass': [],
    'vol': [],
    'density': [],
    'gravity': [],
    'escape': [],
    'meanRadius': [],
    'equaRadius': [],
    'polarRadius': [],
    'flattening': [],
    'sideralOrbit': [],
    'sideralRotation': [],
    'avgTemp': []
}

for planet_info in planet_data.values():
    planets["name"].append(planet_info.get("englishName", None))
    planets['semimajorAxis'].append(planet_info.get("semimajorAxis", None))
    planets['perihelion'].append(planet_info.get("perihelion", None))
    planets['aphelion'].append(planet_info.get("aphelion", None))
    planets['eccentricity'].append(planet_info.get("eccentricity", None))
    planets['inclination'].append(planet_info.get("inclination", None))
    planets['mass'].append(planet_info.get("mass", {}).get("massValue", None))
    planets['vol'].append(planet_info.get("vol", {}).get("volValue", None))
    planets['density'].append(planet_info.get("density", None))
    planets['gravity'].append(planet_info.get("gravity", None))
    planets['escape'].append(planet_info.get("escape", None))
    planets['meanRadius'].append(planet_info.get("meanRadius", None))
    planets['equaRadius'].append(planet_info.get("equaRadius", None))
    planets['polarRadius'].append(planet_info.get("polarRadius", None))
    planets['flattening'].append(planet_info.get("flattening", None))
    planets['sideralOrbit'].append(planet_info.get("sideralOrbit", None))
    planets['sideralRotation'].append(planet_info.get("sideralRotation", None))
    planets['avgTemp'].append(planet_info.get("avgTemp", None))
```
