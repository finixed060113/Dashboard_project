```python
from dotenv import load_dotenv # para cargar las variables de entorno desde el fichero .env 
import requests
import os  # para acceder a las variables de entorno del sistema operativo 
load_dotenv()
#api_key es el nombre de la variable que almacena el token  en el archivo .env
token = os.getenv("api_key")
 
name = 'saturn'

api_url = 'https://api.api-ninjas.com/v1/planets?name={}'.format(name)
response = requests.get(api_url, headers={'X-Api-Key': token})
if response.status_code == requests.codes.ok:
    print(response.text)
else:
    print("Error:", response.status_code, response.text)
```

```python
def get_api(url):
    '''
    Function to make the API call.
    '''
    res = requests.get(url) 
    if res.status_code == 200:  
        return res.json() 
    else:  
        return f"Error {res.status_code}"

res_api = get_api(url)

```

```python
import requests

planet_names = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune"]

planet_data = {}  # Dictionary to store planet data

for planet_name in planet_names:
    
    api_url = f"https://api.le-systeme-solaire.net/rest/bodies/{planet_name}"
    
    response = requests.get(api_url)
    if response.status_code == 200:
        planet_info = response.json()
        planet_data[planet_name] = planet_info
    else:
        print(f"Error fetching data for {planet_name}")

print(planet_data)  # Display the collected planet data
```
